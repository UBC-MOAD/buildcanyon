### Plot schematics for thesis conclusions

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
import scipy.io
%matplotlib inline
from netCDF4 import Dataset
```

```python
filename = 'bathy_2canyons_Astoria_dims.nc'
bathy_data = Dataset(filename)
```

```python
X = bathy_data['X'][:]
Y = bathy_data['Y'][:]
Dep = bathy_data['Bathymetry'][:]
```

```python
plt.rcParams['contour.negative_linestyle'] = 'solid' # default is to have negative contours with dashed lines

# Plot 3D bathymetry

fig = plt.figure(figsize=(20,8))

# `ax` is a 3D-aware axis instance because of the projection='3d' keyword argument to add_subplot
ax = fig.add_subplot(1, 1, 1, projection='3d')

p = ax.plot_surface(X[:-80,70:-70],Y[:-80,70:-70],-Dep[:-80,70:-70], 
                    rstride=2, cstride=2,alpha=0.75,
                    color='beige', 
                    linewidth=0, antialiased=False)
#cb = fig.colorbar(p, shrink=0.65)
cset = ax.contour(X[:-80,70:-70],Y[:-80,70:-70],-Dep[:-80,70:-70],colors='k', rstride=2, cstride=2,zdir='z')

ax.set_aspect('auto')

ax.view_init(40, -70) # elevation and azimuth
ax.grid('off')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
plt.savefig('two_canyons_3D.png', format='png',bbox_inches='tight')
```
