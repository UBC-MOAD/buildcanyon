## Check the final dimensions of the inserts

```python
import cmocean as cmo
import matplotlib.cm as cm
from matplotlib.colors import LightSource
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from netCDF4 import Dataset
import numpy as np
import seaborn as sns
import time
%matplotlib inline
```

```python
sns.set_style('whitegrid')
sns.set_context('talk')
```

```python
filename =  'tank/BathyInsert03.nc'
dataset = Dataset(filename, 'r')
    
xx = dataset.variables['X']
yy = dataset.variables['Y']
bat = dataset.variables['Bathymetry']
```

```python
np.shape(yy)
```

```python
fig, ax = plt.subplots(1,1)
ax.pcolor(xx ,yy ,bat , cmap='Blues')
ax.plot(xx[0,199], yy[0,199], 'ro', markersize=13)
ax.plot(xx[0,0], yy[0,0], 'go', markersize=13)
ax.plot(xx[199,199], yy[199,199], 'r^', markersize=13)
ax.plot(xx[199,0], yy[199,0], 'g^', markersize=13)

# plot circles radius 0.495, 0.212
ax.plot(np.linspace(0,0.495,1000), (-np.linspace(0,0.495,1000)**2+0.495**2)**0.5, color='orange')
ax.plot(np.linspace(0,0.212,1000), (-np.linspace(0,0.212,1000)**2+0.212**2)**0.5, color='orange')

ax.set_xlim(0.0, 0.5)
ax.set_ylim(0.0, 0.5)
ax.set_aspect('equal')
```

```python
# Distance from circle to circle
dist1 = ((xx[0,199]-xx[0,0])**2 + (yy[0,199]-yy[0,0])**2)**0.5
print(dist1)

# Distance from triangle to triangle
dist2 = ((xx[199,199]-xx[199,0])**2 + (yy[199,199]-yy[199,0])**2)**0.5
print(dist2)
```

Looks good...

### Check spacings

```python
filename =  'tank/BathySpacing01.nc'
dataset = Dataset(filename, 'r')
    
xx_s = dataset.variables['X']
yy_s = dataset.variables['Y']
bat_s = dataset.variables['Bathymetry']
```

```python
fig, ax = plt.subplots(1,1)
ax.pcolor(xx_s,yy_s,bat_s, cmap='Blues')
ax.plot(xx_s[0,199], yy_s[0,199], 'ro', markersize=13)
ax.plot(xx_s[0,0], yy_s[0,0], 'go', markersize=13)
ax.plot(xx_s[199,199], yy_s[199,199], 'r^', markersize=13)
ax.plot(xx_s[199,0], yy_s[199,0], 'g^', markersize=13)

# plot circles radius 0.495, 0.212
ax.plot(np.linspace(0,0.495,1000), (-np.linspace(0,0.495,1000)**2+0.495**2)**0.5, color='orange')
ax.plot(np.linspace(0,0.212,1000), (-np.linspace(0,0.212,1000)**2+0.212**2)**0.5, color='orange')

ax.set_xlim(0.0, 0.5)
ax.set_ylim(0.0, 0.5)
ax.set_aspect('equal')
```

```python
# Distance from circle to circle
dist1_s = ((xx_s[0,199]-xx_s[0,0])**2 + (yy_s[0,199]-yy_s[0,0])**2)**0.5
print(dist1_s)

# Distance from triangle to triangle
dist2_s = ((xx_s[199,199]-xx_s[199,0])**2 + (yy_s[199,199]-yy_s[199,0])**2)**0.5
print(dist2_s)
```
